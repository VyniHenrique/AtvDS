```python

contatos = []

def exibir_cabecalho(titulo):

    print("=========================================")
    print(f"    {titulo.upper()}")
    print("=========================================\n")

def cadastrar_contato():
    exibir_cabecalho("Cadastrar Novo Contato")
    nome = input("Digite o nome do contato: ")
    telefone = input("Digite o telefone do contato: ")
    email = input("Digite o e-mail do contato: ")


    if not contatos:
        novo_id = 1
    else:

        novo_id = contatos[-1]['id'] + 1


    novo_contato = {
        "id": novo_id,
        "nome": nome,
        "telefone": telefone,
        "email": email
    }
    contatos.append(novo_contato)

    print("\nContato cadastrado com sucesso!")

def listar_contatos():
    exibir_cabecalho("Lista de Contatos")

    if not contatos:
        print("Nenhum contato cadastrado.")
    else:

        for contato in contatos:
            print(f"ID: {contato['id']}")
            print(f"  Nome: {contato['nome']}")
            print(f"  Telefone: {contato['telefone']}")
            print(f"  E-mail: {contato['email']}")
            print("-----------------------------------------")

    input("\nPressione Enter para voltar ao menu principal...")

def atualizar_contato():
    exibir_cabecalho("Atualizar Contato")

    if not contatos:
        print("Nenhum contato para atualizar.")
        return

    try:
        id_contato = int(input("Digite o ID do contato que deseja atualizar: "))
        contato_encontrado = None


        for contato in contatos:
            if contato["id"] == id_contato:
                contato_encontrado = contato
                break

        if contato_encontrado:
            print(f"\nAtualizando contato: {contato_encontrado['nome']}\n")
            nome = input(f"Novo nome (deixe em branco para manter '{contato_encontrado['nome']}'): ")
            telefone = input(f"Novo telefone (deixe em branco para manter '{contato_encontrado['telefone']}'): ")
            email = input(f"Novo e-mail (deixe em branco para manter '{contato_encontrado['email']}'): ")

            if nome:
                contato_encontrado['nome'] = nome
            if telefone:
                contato_encontrado['telefone'] = telefone
            if email:
                contato_encontrado['email'] = email

            print("\nContato atualizado com sucesso!")
        else:
            print("\nID de contato não encontrado.")

    except ValueError:
        print("\nEntrada inválida. Por favor, digite um número de ID.")



def excluir_contato():
    exibir_cabecalho("Excluir Contato")

    if not contatos:
        print("Nenhum contato para excluir.")

        return

    try:
        id_contato = int(input("Digite o ID do contato que deseja excluir: "))
        contato_encontrado = None


        for contato in contatos:
            if contato["id"] == id_contato:
                contato_encontrado = contato
                break

        if contato_encontrado:

            contatos.remove(contato_encontrado)
            print(f"\nContato '{contato_encontrado['nome']}' excluído com sucesso!")
        else:
            print("\nID de contato não encontrado.")

    except ValueError:
        print("\nEntrada inválida. Por favor, digite um número de ID.")



def menu_principal():
    while True:
        exibir_cabecalho("Menu Principal - Agenda com Dicionários")
        print("1. Cadastrar Contato (Create)")
        print("2. Listar Contatos (Read)")
        print("3. Atualizar Contato (Update)")
        print("4. Excluir Contato (Delete)")
        print("5. Sair")
        print("\n=========================================")

        opcao = input("Escolha uma opção: ")

        if opcao == '1':
            cadastrar_contato()
        elif opcao == '2':
            listar_contatos()
        elif opcao == '3':
            atualizar_contato()
        elif opcao == '4':
            excluir_contato()
        elif opcao == '5':
            print("Saindo do programa. Até mais!")
            break
        else:
            print("\nOpção inválida. Por favor, tente novamente.")

if __name__ == "__main__":
    menu_principal()
```
