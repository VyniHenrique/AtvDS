```python
import kagglehub
from kagglehub import KaggleDatasetAdapter
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
file_path = '/content/USvideos.csv'

df = pd.read_csv(file_path)
coluna_dados = df['publish_time']
```

```python
df['publish_time'] = pd.to_datetime(df['publish_time'])
```

Now that the `publish_time` column is in datetime format, we can create a histogram based on different aspects of the date, such as the year or month. Would you like to see a histogram of the publication years or months?

```python
plt.hist(df['publish_time'].dt.year, bins=df['publish_time'].dt.year.nunique(), edgecolor='black')
plt.title("Distribution of Publication Years")
plt.xlabel("Year")
plt.ylabel("Number of Videos")
plt.show()
```

```python
plt.hist(df['publish_time'].dt.hour, bins=24, edgecolor='black')
plt.title("Distribution of Publication Hours")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Videos")
plt.xticks(range(0, 24))
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df['publish_time'].dt.hour, fill=True)
plt.title("Kernel Density Estimation of Publication Hours")
plt.xlabel("Hour of Day")
plt.ylabel("Density")
plt.xticks(range(0, 24))
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['publish_time'].dt.hour)
plt.title("Boxplot of Publication Hours Without Outliers")
plt.xlabel("Hour of Day")
plt.show()
```
